```python
import spacy
```

```python

text = "The children love cream biscuits"
nlp = spacy.load("en_core_web_sm")
# create spacy 
doc = nlp(text)

for token in doc:
    print(token.text,'->',token.pos_)
```

```python
text = "Hellow world how are you , i think people have problems in understanding the NLP"
nlp = spacy.load("en_core_web_sm")
doc = nlp(text)
for token in doc:
    # check token pos
    if token.pos_=='NOUN':
        # print token
        print(token.text)
```

```python
from spacy import displacy 
displacy.render(doc, style='dep',jupyter=True)
```

```python
for token in doc:
    # extract subject
    if (token.dep_=='nsubj'):
        print(token.text)
    # extract object
    elif (token.dep_=='dobj'):
        print(token.text)
```
